```python
import gd
import numpy as np
import matplotlib.pyplot as plt
```

# Simple Projective Problem

```python
n=10
q_0 = np.ones(n)/n
losses = np.random.rand(10)
loss_func = lambda q: np.dot(q,losses)
```

```python
g = gd.projective_simplex_gradient_descent(q_0,loss_func,eta=0.1)
```

```python
plt.plot(g)
plt.plot(losses)
```

```python
def decider(t,sorted_x):
    t_hat = None
    for i in range(len(t)-1):
        if t[i]>=sorted_x[::-1][i+1]:
            return t[i]
    t_hat = t[-1]
    return t_hat
```
