# Head to head

```python
def decider(t,sorted_x):
    t_hat = None
    for i in range(len(t)-1):
        if t[i]>=sorted_x[::-1][i+1]:
            return t[i]
    t_hat = t[-1]
    return t_hat

def project_to_simplex_2(x):
    sorted_x = sorted(x)
    t = [np.sum((np.array(sorted_x[-i:])-1)/i) for i in range(1,len(sorted_x)+1)]
    t_hat = decider(t,sorted_x)
    output = gd._plus(x-decider(t,sorted_x))
    output/=np.sum(output)
    
    return output
```

```python
x = np.random.randn(10)
```

```python
x_0 = gd.project_to_simplex(x)
```

```python
y_0 = project_to_simplex_2(x)
```

```python
plt.plot(x)
plt.plot(x_0,label="Projection 1")
plt.plot(y_0,label="Projection 2")
plt.legend()
```

```python
np.linalg.norm(x_0-x)
```

```python
np.linalg.norm(y_0-x)
```
