```python
actual_values = np.random.randint(0,10,size=[40])
```

```python
def create_expert(std,actual_values):
    return actual_values+np.random.randn(actual_values.size)*std

# Fixed Standard Deviations for each expert
stds = [1,0.1,0.03,0.5,1.3,2,5,1,2,5]
expert_predictions = np.array([create_expert(std,actual_values) for std in stds])
```

We consider the example case where we have a collection of experts where the mean of their predictions is the true mean, but they all have varied standard deviations. Therefore an optimal algorithm will concentrate the probability on the third element.

```python
from scipy.optimize import nnls
```

```python
weights, loss = nnls(expert_predictions.T, actual_values)
```

```python
plt.plot(weights)
```

```python
plt.plot(np.dot(weights,expert_predictions))
plt.plot(actual_values)
```
